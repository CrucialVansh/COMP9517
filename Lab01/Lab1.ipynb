```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Task 1

```python
image = cv.imread("Task1.jpg")
invert = 255 - image
gamma_bg = 1
threshold = 500
normalized = image / 255.0
final_image = np.where(invert < threshold, invert, np.power(normalized, gamma_bg) * 255).astype(np.uint8) 
_, axs = plt.subplots(1,2, figsize=(15, 5))
axs[0].imshow(image)
axs[0].set_title("Input Image")
axs[0].axis("off")
axs[1].axis("off")
axs[1].set_title("Output Image")
axs[1].imshow(final_image)
plt.show()
```

# Task 2

```python
panda_image = cv.imread("Task2.jpg")
filtered = cv.boxFilter(panda_image, -1, (9, 9))
filtered1 = cv.GaussianBlur(panda_image, (9, 9), 5)
filtered2 = cv.medianBlur(panda_image, 3)
_, axs = plt.subplots(1,2, figsize=(15, 5))
axs[0].imshow(panda_image)
axs[0].axis("off")
axs[0].set_title("Input Image")
axs[1].imshow(filtered2)
axs[1].axis("off")
axs[1].set_title("Output Image")
```

I used the median blur to get rid of the noise. As discussed in the course, median blur is best if there is a need to remove isolated
intensity spikes. Median makes it so pixels are similar to their neighbours. This is especially useful in this cases as there is a lot of salt and pepper noise. Median filter produces the best result compared to Normal or Gaussian blur as they are too smoothed and not sharp.  

# Task 3

```python
blurred_image = cv.imread("Task3.jpg", cv.IMREAD_GRAYSCALE)
normal = cv.normalize(blurred_image, None, alpha=0, beta=255, norm_type=cv.NORM_MINMAX)
blurred = cv.GaussianBlur(normal, (3, 3), 0)
high = blurred_image - blurred 
sharpened = cv.addWeighted(blurred_image, 1, high, -0.3, 0)
laplacian = cv.Laplacian(normal, cv.CV_64F, 1)
laplacian_abs = np.uint8(np.absolute(laplacian))
lap = cv.subtract(blurred_image, laplacian_abs)
_, axs = plt.subplots(1,3, figsize=(15, 5))
axs[0].imshow(blurred_image, cmap='gray')
axs[0].set_title("Input Image")
axs[1].set_title("Unsharp Mask")
axs[1].imshow(sharpened, cmap='gray')
axs[0].axis("off")
axs[1].axis("off")
axs[2].imshow(lap, cmap='gray')
axs[2].axis("off")
axs[2].set_title("Laplacian")
cv.imshow("g", shar)
cv.waitKey(0)
```
