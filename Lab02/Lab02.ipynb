```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Task 1

```python
image1 = cv.imread("Image1.jpg", cv.IMREAD_GRAYSCALE)
temp = cv.imread("Image1.jpg", cv.IMREAD_COLOR)
rgbI = cv.cvtColor(temp, cv.COLOR_BGR2RGB)
sift = cv.SIFT_create(20, 3, 0.1)
keypoints1, des1 = sift.detectAndCompute(image1, None)
# Copy the original image to draw on it
custom_image = rgbI.copy()

# Loop through each keypoint and draw a circle with custom radius and thickness
for kp in keypoints1:
    # Get the center coordinates of the keypoint
    x, y = int(kp.pt[0]), int(kp.pt[1])
    # You can use kp.size to derive a radius, or define a custom value
    radius = int(kp.size / 2)  # or any custom value, e.g., radius = 10
    thickness = 2  # Set your desired thickness here
    # Draw the circle on the image (using green color as an example)
    cv.circle(custom_image, (x, y), radius, (0, 255, 0), thickness)

# Display the image with custom keypoints
plt.imshow(custom_image)
plt.axis("off")
plt.show()
```
